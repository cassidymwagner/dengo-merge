```python
import os
os.environ['YT_DEST'] = '/home/kwoksun2/anaconda2/pkgs/yt-3.3.5-np111py27_2/'
import numpy as np
from dengo.chemical_network import \
    ChemicalNetwork, \
    reaction_registry, \
    cooling_registry, species_registry
import dengo.primordial_rates, dengo.primordial_cooling
from dengo.chemistry_constants import tiny, kboltz, mh
import yt
```

```python
NCELLS = 10.0
density = 1e8
temperature = np.logspace(2, 4, NCELLS)


NCELLS = 100.0
temp_ = np.logspace(2, 3.5, 10)
den_  = np.logspace(7, 10 , 10)
temperature = np.array([ temp_ for i in den_  ]).flatten()
density  = np.array([ den_  for i in temp_ ]).transpose().flatten()


X = 0.75

# initialize and setup the network
dengo.primordial_rates.setup_primordial()
primordial = ChemicalNetwork()
```

```python
for i in range(23):
    try: 
        primordial.add_reaction("k{0:02d}".format(i+1))
    except:
        pass

# This defines the temperature range for the rate tables
# set e.g. logT, T
primordial.init_temperature((1e0, 1e8))

tiny = 1e-10


# everything is in mass density
init_array = density
init_values = dict()

init_values['H_2']     = X * init_array
init_values["H_1"]     = X * init_array
init_values["H_m0"]     = X * init_array

init_values["H2"]     = X * init_array*2.
init_values["H2_1"]     = X * init_array*2.
init_values["H2_2"]     = X * init_array*2.

init_values["He_1"] = ( 1.0 - X )*init_array*2.
init_values["He_2"] = ( 1.0 - X )*init_array*2.
init_values["He_3"] = ( 1.0 - X )*init_array*2.

total_density = primordial.calculate_total_density(init_values)


init_values = primordial.convert_to_mass_density(init_values)
init_values['de'] = primordial.calculate_free_electrons(init_values)
init_values['density'] = primordial.calculate_total_density(init_values)
number_density = primordial.calculate_number_density(init_values)

# set up initial temperatures values used to define ge
init_values['T'] = temperature

# calculate ge (very crudely, no H2 help here)
gamma = 5.0/3.0
init_values['ge'] = ((temperature * number_density * kboltz)
                     / (init_values['density'] * mh * (gamma - 1)))

primordial.write_solver("sunny", output_dir = ".",
                        init_values=init_values,
                        input_is_number=False)
```

```python
import pyximport
pyximport.install(setup_args={"include_dirs":np.get_include()},
                  reload_support=True, inplace=True)

sunny_solver_run = pyximport.load_module("sunny_solver_run",
                            "sunny_solver_run.pyx",
                            build_inplace = True, pyxbuild_dir = "_dengo_temp")
rv, rv_int = sunny_solver_run.run_sunny(init_values, 1e10, niter=1e5)
import pylab
import matplotlib.pyplot as plt
plt.clf()
```

```python
mask = rv_int['successful']
for name in sorted(rv_int):
    if len(rv_int[name].shape) == 1:
        rv_int[name] = rv_int[name][mask]
    else:
        rv_int[name] = rv_int[name][0, mask]

skip = ('successful', 'dt', 't', 'ge', 'He_1', 'He_2', 'He_3')
for n, v in sorted(rv_int.items()):
    if n in skip: continue
    plt.loglog(rv_int['t'], v, label = n)


plt.xlabel("time [s]")
plt.legend(loc='best',fontsize = 'small')
plt.ylim(1e-4,1e11)
plt.savefig("6-species.png")


plt.clf()
plt.loglog(rv_int['t'], rv_int['T'], label = 'T')

plt.xlabel("time [s]")
plt.savefig("plot_temp-6-species.png")
```

```python
def equil_values(temperature, density):
#     NCELLS = 10.0
#     density = 1e8
#     temperature = np.logspace(2, 4, NCELLS)


#     NCELLS = 100.0
#     temp_ = np.logspace(2, 3.5, 10)
#     den_  = np.logspace(7, 10 , 10)
#     temperature = np.array([ temp_ for i in den_  ]).flatten()
#     density  = np.array([ den_  for i in temp_ ]).transpose().flatten()
    
    NCELLS = len(temperature)

    X = 0.75

    # initialize and setup the network
    dengo.primordial_rates.setup_primordial()
    primordial = ChemicalNetwork()

    for i in range(23):
        try: 
            primordial.add_reaction("k{0:02d}".format(i+1))
        except:
            pass

    # This defines the temperature range for the rate tables
    # set e.g. logT, T
    primordial.init_temperature((1e0, 1e8))

    tiny = 1e-10


    # everything is in mass density
    init_array = np.ones(NCELLS)*density
    init_values = dict()

    init_values['H_2']     = X * init_array
    init_values["H_1"]     = X * init_array
    init_values["H_m0"]     = X * init_array

    init_values["H2"]     = tiny * init_array*2.
    init_values["H2_1"]     = tiny * init_array*2.
    init_values["H2_2"]     = tiny * init_array*2.

    init_values["He_1"] = ( 1.0 - X )*init_array*2.
    init_values["He_2"] = ( 1.0 - X )*tiny*init_array*2.
    init_values["He_3"] = ( 1.0 - X )*init_array*2.

    total_density = primordial.calculate_total_density(init_values)


    init_values = primordial.convert_to_mass_density(init_values)
    init_values['de'] = primordial.calculate_free_electrons(init_values)
    init_values['density'] = primordial.calculate_total_density(init_values)
    number_density = primordial.calculate_number_density(init_values)

    # set up initial temperatures values used to define ge
    init_values['T'] = temperature

    # calculate ge (very crudely, no H2 help here)
    gamma = 5.0/3.0
    init_values['ge'] = ((temperature * number_density * kboltz)
                         / (init_values['density'] * mh * (gamma - 1)))
    
    !rm sunny*
    
    primordial.write_solver("sunny", output_dir = ".",
                            init_values=init_values,
                            input_is_number=False)

    sunny_solver_run = pyximport.load_module("sunny_solver_run",
                                "sunny_solver_run.pyx",
                                build_inplace = True, pyxbuild_dir = "_dengo_temp")
    rv, rv_int = sunny_solver_run.run_sunny(init_values, 3.16e6, niter=1e5) ;
    
    mask = rv_int['successful']
    
    return rv_int['H_1'][:, mask][:,-1] / init_values['density'], rv_int['t'][mask][-1]
```

```python
import numpy

dendim = 64
tempdim = 64

temp = numpy.linspace(1000 , 10000  , tempdim)
den = numpy.logspace(12, 20, dendim)

#den = numpy.array([1.0e4, 1.0e10])
#temp = numpy.array([100.0,5000.0])

H1_array = numpy.zeros((dendim, tempdim))
t_array = numpy.zeros((dendim))

for i, iden in enumerate(den):
    H1_, t = equil_values(temp, iden)
    print(H1_)
    t_array[i] = t
    H1_array[i,:] = H1_
```

```python
import matplotlib.pyplot as plt
def get_species_phase_map(species):
    change_in_H2 = []
    den_ratio = []
    

    for i, init_den in enumerate(den):
        mask = (H2_array[i]['successful'])
        if species != 't':
            a = np.array(H2_array[i][species][:, mask][:,-1])
            den_ratio.append(a/init_den)
            
        else:
            a = np.array(H2_array[i][species][mask][-1])
            den_ratio.append(a)
            
    return numpy.array(den_ratio)

def get_species_abundance(species):
    change_in_H2 = []
    den_ratio = []
    

    for i, init_den in enumerate(den):
        mask = (H2_array[i]['successful'])
        if species != 't':
            a = np.array(H2_array[i][species][:, mask][:,-1])
            den_ratio.append(a)
            
        else:
            a = np.array(H2_array[i][species][mask][-1])
            den_ratio.append(a)
            
    return numpy.array(den_ratio)


im = get_species_phase_map('H2_1')

plt.imshow(im,origin='lower')
plt.xlabel('Temperature ($10^{2} - 10^{4}$ K)')
plt.ylabel('H2_1 ($10^{1} - 10^{15} cm^{-3}$)')
plt.colorbar()
plt.show()

im = get_species_phase_map('t')
print(im/3.2e10)


im = get_species_abundance('H2_1')
plt.imshow(numpy.log10(im),origin='lower')
plt.colorbar()
plt.show()
```

```python
k13_reaction = primordial.reactions['k13']
k22_reaction = primordial.reactions['k22']
```

```python
k13 = (k13_reaction.coeff_fn(primordial))
k22 = (k22_reaction.coeff_fn(primordial))
```

```python
def f_H_eq(density): 
    return (numpy.sqrt(8.0*k13*k22*density+k13*k13) - k13 )/4.0/k22/density
```

```python
fh_array = numpy.zeros((len(den), 1024))

for i, iden in enumerate(den):
#    print(i)
    fh_array[i,:] = f_H_eq(iden)
```

```python
import matplotlib.pyplot as plt

fh_array[fh_array<1e-10] = 1e-10

X,Y = numpy.meshgrid(den, primordial.T)

# fh_array[fh_array>0] = 0.0

plt.clf()
imag = plt.contourf( X, Y, numpy.log10(fh_array).transpose(), 11 )
plt.xscale('log')
# plt.yscale('log')
#plt.xticks(den)
#plt.yticks(temp)
plt.colorbar()
plt.show()
```
