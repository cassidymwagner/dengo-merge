```python
species_list = ["H2_1","H2_2","H_1","H_2","H_m0",
                "He_1","He_2","He_3","de","ge"]
```

```python
def get_error(name):
    error_list = []
    tol_list = [] # reltol solution
    sol_list = [] # reference solution
    
    for i in range(1024):
        j = i + 1
        with open( "error/{}_error.out".format(j) ) as f:
            for line in f.readlines():
                if name in line:
                    error, tol, sol = line.strip('\n').split(':')[1].split(';')
                    
                    error_list.append(float(error))
                    tol_list.append(float(tol))
                    sol_list.append(float(sol))
    return numpy.array(error_list), numpy.array(tol_list), numpy.array(sol_list)
```

```python
error_dict = {}
for sp  in species_list:
    error_dict[sp] = get_error(sp)
    plt.semilogy(error_dict[sp][0],label=sp)
    if (error_dict[sp][0] < 0.1 ).sum() <= 1024 :
        print(sp,  (error_dict[sp][0] > 0.1).sum() )
plt.ylim(1e-16,1e10)
plt.legend()
```

```python
for sp in species_list:
    make_2dcolorplot( error_dict[sp][0].reshape((32,32)) ,  title = sp, vmin=-2, vmax=2, cmap='RdYlBu', cbar_label='log(fractional error)' )
```

```python
log_bins = numpy.logspace(-2,2, 100)
for sp in species_list:
    error_array = numpy.clip(error_dict[sp][0], 1.0e-2,1.0e2)
    plt.hist( error_array , label=sp, bins = log_bins, alpha = 0.5 )
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
def get_by_name( line, name, update_list ):
    if name in line:
        update_list.append( int(line.split(':')[1]) )
        return update_list
    else:
        return update_list
    
def get_stats_for_given_file(filename, list_of_stats = ['nsteps', 'nfevals', 'nlinsetups', 'netfails', 'njevals','nncfails','nniters'] ) :
    tsteps = []

    stats_dict = {}
    for attr in list_of_stats:
        stats_dict[attr] = []

    with open(filename) as f:
        for line in f.readlines():
            if "iterations" in line:
                _, niter, tnow = (line.split(':'))
                tnow = float(tnow.strip('\n'))
                tsteps.append(tnow)
            for attr in list_of_stats:
                stats_dict[attr] = get_by_name( line, attr, stats_dict[attr] )

    for attr in list_of_stats:
        stats_dict[attr] = numpy.array(stats_dict[attr])

    tsteps = numpy.array(tsteps)
    dt = tsteps[1:] - tsteps[:-1]
    dt = numpy.append(numpy.clip(dt,0,1),1)
    success = numpy.array(dt, dtype=bool)
    
    frac_stats_dict = {}
    for attr in list_of_stats:
        fail_stats = float(stats_dict[attr][~success].sum())
        total_stats = float(stats_dict[attr].sum())
        if total_stats != 0:
            frac_attr = fail_stats / total_stats 
        else:
            frac_attr = 0.0
        frac_stats_dict[attr] = frac_attr
    return frac_stats_dict
```

```python
all_frac_stats = {}
list_of_stats = ['nsteps', 'nfevals', 'nlinsetups', 'netfails', 'njevals','nncfails','nniters']

for attr in list_of_stats:
    all_frac_stats[attr] = []

for i in range(1024):

    stats_dict = get_stats_for_given_file('stats/{}_stats.out'.format(i + 1))
    for k, v in stats_dict.items():
        all_frac_stats[k].append(v)
for attr in list_of_stats:
    all_frac_stats[attr] = numpy.array(all_frac_stats[attr])
```

```python
for attr in list_of_stats:
    make_2dcolorplot( all_frac_stats[attr].reshape((32,32)) ,  title = attr, vmin=0, vmax=1, log_z = False )
```

```python
def get_name_from_file( name = "user", filename = 'runtest.output' ):
    f = open(filename, 'r')
    result = list()
    for line in f:
        if name in line:
            result.append(line)
    f.close()
    return result

def change_raw_time_to_time(raw_time):
    return raw_time[0].split('\t')[1].strip('\n')

def change_time_to_s( min_s ):
    minutes, seconds =  min_s.strip('s').split('m')
    minute = float(minutes)
    second = float(seconds)
    
    total_time = minute*60.0 + second
    
    return total_time
```

```python
list_time = []
for i in range(1024):
    raw_time = get_name_from_file(name = 'user', filename = 'stats/{}_stats.out'.format(i+1))
    time = change_raw_time_to_time(raw_time)
    time = change_time_to_s(time)
    list_time.append(time)
list_time = numpy.array(list_time)
```

```python
make_2dcolorplot( list_time.reshape((32,32)) / 1024, 
                 vmin = -5, vmax = -2 ,
                 title = 'rel=1e-3 runtime', cbar_label='log(runtime)' )
```
