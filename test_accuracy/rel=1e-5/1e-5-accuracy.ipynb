```python
species_list = ["H2_1","H2_2","H_1","H_2","H_m0",
                "He_1","He_2","He_3","de","ge", "T"]
```

```python
def get_error(name):
    error_list = []
    tol_list = [] # reltol solution
    sol_list = [] # reference solution
    
    for i in range(1024):
        j = i + 1
        with open( "error/{}_error.out".format(j) ) as f:
            for line in f.readlines():
                if name in line:
                    error, tol, sol = line.strip('\n').split(':')[1].split(';')
                    
                    error_list.append(float(error))
                    tol_list.append(float(tol))
                    sol_list.append(float(sol))
    return numpy.array(error_list), numpy.array(tol_list), numpy.array(sol_list)
```

```python
error_dict = {}
for sp  in species_list:
    error_dict[sp] = get_error(sp)
    plt.semilogy(error_dict[sp][0],label=sp)
    if (error_dict[sp][0] < 0.1 ).sum() <= 1024 :
        print(sp,  (error_dict[sp][0] > 0.1).sum() )

plt.ylim(1e-16,1e10)
plt.legend()
```

```python
rhosp_dict0 = {}
rhosp_dict1 = {}
for sp  in species_list:
    rhosp_dict0[sp] = get_error(sp)[1]
    rhosp_dict1[sp] = get_error(sp)[2]

mass_dict = {'H_2': 1.00794, 'H_1': 1.00794, 'de': 1.0, 'He_1': 4.002602, 
 'He_2': 4.002602, 'He_3': 4.002602, 'ge': 1.0, 'H2_2': 2.01588, 'H2_1': 2.01588, 'H_m0': 1.00794}

charge_dict = {'H2_1': 0.0,'H2_2': 1.0,'H_1': 0.0,'H_2': 1.0, 'H_m0': -1.0, 'He_1': 0.0,'He_2': 1.0,'He_3': 2.0,'de': 0.0}

def calculate_H_density( mass_density_dict):
    total_H = numpy.zeros(len(mass_density_dict['ge']))
    for k in ['H_2','H_1','H2_1','H2_2','H_m0']:
        total_H += mass_density_dict[k]
    return total_H

def calculate_He_density( mass_density_dict):
    total_H = numpy.zeros(len(mass_density_dict['ge']))
    for k in ['He_1','He_2','He_3']:
        total_H += mass_density_dict[k]
    return total_H

def convert_to_number_density( mass_density_dict ):
    num_dict = {}
    for k , v in mass_dict.items():
        num_dict[k] = mass_density_dict[k] / mass_dict[k]
    return num_dict

def get_electron_density( mass_density_dict ):
    num_den_dict = convert_to_number_density( mass_density_dict )
    total_e = numpy.zeros(len(mass_density_dict['ge']))
    for k, v in charge_dict.items():
        total_e += num_den_dict[k] * v
    
    conserve_efrac = (total_e - num_den_dict['de']) / num_den_dict['de']
    return conserve_efrac

    
```

```python
rhoH0 = calculate_H_density(rhosp_dict0)
rhoH1 = calculate_H_density(rhosp_dict1)

rhoHe0 = calculate_He_density(rhosp_dict0)
rhoHe1 = calculate_He_density(rhosp_dict1)

efrac0 = get_electron_density(rhosp_dict0)
efrac1 = get_electron_density(rhosp_dict1)

plt.title('fractional conservation of mass')
plt.plot(rhoH0/rhoH1, label='H')
plt.plot(rhoHe0/rhoHe1, label='He')
plt.legend()
plt.show()
plt.title('fractional conservation of de')
plt.plot(efrac0, label='de0')
plt.plot(efrac1, label='de1')
plt.legend()
plt.show()
```

```python
def get_by_name( line, name, update_list ):
    if name in line:
        update_list.append( int(line.split(':')[1]) )
        return update_list
    else:
        return update_list
    
def get_stats_for_given_file(filename, list_of_stats = ['nsteps', 'nfevals', 'nlinsetups', 'netfails', 'njevals','nncfails','nniters'] ) :
    tsteps = []

    stats_dict = {}
    for attr in list_of_stats:
        stats_dict[attr] = []

    with open(filename) as f:
        for line in f.readlines():
            if "iterations" in line:
                _, niter, tnow = (line.split(':'))
                tnow = float(tnow.strip('\n'))
                tsteps.append(tnow)
            for attr in list_of_stats:
                stats_dict[attr] = get_by_name( line, attr, stats_dict[attr] )

    for attr in list_of_stats:
        stats_dict[attr] = numpy.array(stats_dict[attr])

    tsteps = numpy.array(tsteps)
    dt = tsteps[1:] - tsteps[:-1]
    dt = numpy.append(numpy.clip(dt,0,1),1)
    success = numpy.array(dt, dtype=bool)
    
    frac_stats_dict = {}
    for attr in list_of_stats:
        fail_stats = float(stats_dict[attr][~success].sum())
        total_stats = float(stats_dict[attr].sum())
        if total_stats != 0:
            frac_attr = fail_stats / total_stats 
        else:
            frac_attr = 0.0
        frac_stats_dict[attr] = frac_attr
    return frac_stats_dict
```

```python
all_frac_stats = {}
list_of_stats = ['nsteps', 'nfevals', 'nlinsetups', 'netfails', 'njevals','nncfails','nniters']

for attr in list_of_stats:
    all_frac_stats[attr] = []

for i in range(1024):

    stats_dict = get_stats_for_given_file('stats/{}_stats.out'.format(i + 1))
    for k, v in stats_dict.items():
        all_frac_stats[k].append(v)
for attr in list_of_stats:
    all_frac_stats[attr] = numpy.array(all_frac_stats[attr])
```

```python
for attr in list_of_stats:
    make_2dcolorplot( all_frac_stats[attr].reshape((32,32)) ,  title = attr, vmin=0, vmax=1, log_z = False, cbar_label='fraction of failed iteration' )
```

```python
def get_name_from_file( name = "user", filename = 'runtest.output' ):
    f = open(filename, 'r')
    result = list()
    for line in f:
        if name in line:
            result.append(line)
    f.close()
    return result

def change_raw_time_to_time(raw_time):
    return raw_time[0].split('\t')[1].strip('\n')

def change_time_to_s( min_s ):
    minutes, seconds =  min_s.strip('s').split('m')
    minute = float(minutes)
    second = float(seconds)
    
    total_time = minute*60.0 + second
    
    return total_time
```

```python
list_time = []
for i in range(1024):
    raw_time = get_name_from_file(name = 'user', filename = 'stats/{}_stats.out'.format(i+1))
    time = change_raw_time_to_time(raw_time)
    time = change_time_to_s(time)
    list_time.append(time)
list_time = numpy.array(list_time)
```
