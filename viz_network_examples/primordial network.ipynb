```python
import networkx as nx
import matplotlib.pyplot as plt

from dengo import chemical_network
from dengo.chemical_network import \
    ChemicalNetwork, \
    reaction_registry, \
    cooling_registry, species_registry
import dengo.primordial_rates, dengo.primordial_cooling
from dengo.get_rates import setup_umist_species, setup_umist_reactions
```

```python
dengo.primordial_rates.setup_primordial()
primordial = ChemicalNetwork()

for i in range(23):    
    try:
        primordial.add_reaction("k{0:02d}".format(i+1))
    except:
        pass
```

```python
G = nx.MultiDiGraph()

for reaction in primordial.reactions:
    r = primordial.reactions[reaction]
    print(r)
    down_species  = r.down_species
    up_species  = r.up_species    
    for d_sp in down_species:
        for u_sp in up_species:
            G.add_edge( d_sp.name, u_sp.name )
```

```python
nx.shell_layout(G)
p = nx.draw(G, with_labels=True, arrows=True, node_size=1000)
```

```python
import snakes.plugins
snakes.plugins.load("gv", "snakes.nets", "nets")
from nets import *

def chem_network () :
    # initialize the net
    n = PetriNet("N")
    
    for sp in primordial.required_species:
        # add the nodes to the network
        n.add_place(Place(sp.name, []))
    
    primordial_reactions = primordial.reactions
    for reaction in primordial_reactions:
        r = primordial_reactions[reaction]
        t = Transition(reaction)
        n.add_transition(t)
        
        left_side = r.left_side
        right_side = r.right_side
        
        if reaction not in ["k22","k23"]:
            print(reaction)
            for sp in left_side:
                n.add_input( sp[1].name , reaction, Value(sp[0]))
            for sp in right_side:    
                n.add_output(sp[1].name , reaction, Value(sp[0]))
        else:
            n.add_input( sp[1].name, reaction, Value(3) )
            for sp in right_side:    
                n.add_output(sp[1].name , reaction, Value(sp[0]))
            
    return n, t, t.modes()
```

```python
net, trans, modes = chem_network()
```
